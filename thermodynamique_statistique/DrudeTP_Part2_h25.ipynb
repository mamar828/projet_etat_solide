### Question statistique ###

VIII. _(5 points)_ &mdash; Vérifiez numériquement et graphiquement que l'amortissement de la quantité de mouvement moyenne des électrons suit l'équation exponentielle dérivée analytiquement en classe, soit $\langle p(t)\rangle =\langle p(t_0)\rangle \,e^{-t/\tau}$, et comparez-y le comportement d'un seul électron.

On ajuste ici une variante de l'équation exponentielle afin de tenir compte d'un décalage vertical provenant d'une température constante.

```python
complete_p = np.loadtxt("thermodynamique_statistique/data/drude_p.csv", delimiter=",")
mean_p = complete_p.mean(axis=1)
t = np.arange(complete_p.shape[0])

fig = gl.Figure(size=(11,7), x_label="Temps [u. arb.]", y_label="Quantité de mouvement [u. arb.]", figure_style="dim")
mean_p_curve = gl.Curve(t, mean_p, line_width=2, label="Moyenne des électrons", color="yellow")
single_p_curve = gl.Curve(t, complete_p[:,3], line_width=1, label="Électron unique", color="blue")
fit = gl.FitFromFunction(
    lambda t, tau, k: (mean_p[0] - k) * np.exp(-t / tau) + k,
    mean_p_curve,
    guesses=[10000,3.2e-25],
    color="red",
    line_style=":",
    line_width=2
)
fit.label = rf"Amortissement ajusté, $\tau={fit.parameters[0]:.2f}$"

fig.add_elements(single_p_curve, mean_p_curve, fit)
fig.show()
```

Figure 1. Quantité de mouvement en fonction du temps dans le modèle de Drude. La ligne rouge représente un ajustement exponentiel sur la position moyenne des électrons en utilisant une équation de la forme $\left<p(t)\right>=\left(\left<p(t_0)-k\right>\right)e^{-t/\tau}+k$, dont le paramètre de translation verticale résulte d'une température constante. Tel que discuté avec la professeure, une température constante est acceptée.

X. _(5 points)_ &mdash; Pour quelques différents modules de champ électrique, présentez graphiquement l'évolution de la position moyenne des électrons en fonction du temps pour ses deux composantes parallèle et perpendiculaire au champ.

```python
data = {
    str(magnitude) : np.loadtxt(f"thermodynamique_statistique/data/drude_apos_E={magnitude}.csv", delimiter=",")
    for magnitude in [1e-22, 2e-22, 3e-22, 5e-22, 7e-22]
}
t = np.arange(data["1e-22"].shape[0])

fig = gl.Figure(size=(11,7), x_label="Temps [u. arb.]", y_label="Position moyenne [u. arb.]", figure_style="dim")
curves = []
for (magnitude, data_array), color in zip(data.items(), ["purple", "pink", "lime", "cyan", "orange"]):
    curves.append(gl.Curve(t, data_array[:,::2].mean(axis=1), color=color, line_style="-",
                           label=f"$E_y={magnitude[0]}"+r"\cdot10^{-23}$"))
    curves.append(gl.Curve(t, data_array[:,1::2].mean(axis=1), color=color, line_style=":"))

fig.add_elements(*curves)
fig.show()
```

Figure 2. Position moyenne des électrons en fonction du temps pour cinq modules de champ électrique différents. Les lignes pleines représentent la composante perpendiculaire au champ alors que les lignes pointillées correspondent à la composante parallèle au champ. Le module du champ, en unités arbitraires, est écrit en légende pour chaque courbe.
